# Stage graphs, disaggregation, and connectors


vLLM-Omni represents heterogeneous pipelines as stages. Each stage can have its own engine, replicas, devices, and batching policy. Connectors move intermediate payloads.

```python
from omni_tutorial import build_voice_pipeline
graph=build_voice_pipeline()
output,trace=graph.run("Describe the image")
for event in trace: print(event["stage"],event["kind"],"=>",event["output"])
print("final",output,"transfers",graph.connector.transfers)
```

The payload evolves from semantic text/hidden state, to audio codec tokens, to waveform metadata. In a real system, a connector may transport tensors or KV caches via shared memory or a network transfer engine.
