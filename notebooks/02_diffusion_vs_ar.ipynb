# Diffusion versus autoregressive generation


Diffusion begins with noise and iteratively moves a latent toward a conditioned target. Unlike AR decoding, one step updates a whole latent array rather than appending one token.

```python
import numpy as np
import matplotlib.pyplot as plt
rng=np.random.default_rng(7)
target=np.array([2.0,-1.0])
x=rng.normal(size=2)*4
trajectory=[x.copy()]
for step in range(20):
    x=x+0.18*(target-x)+rng.normal(scale=0.08,size=2)
    trajectory.append(x.copy())
trajectory=np.array(trajectory)
plt.plot(trajectory[:,0],trajectory[:,1],marker="o")
plt.scatter(*target,color="red",label="conditioned target")
plt.legend(); plt.title("Toy denoising trajectory"); plt.show()
print("final distance",np.linalg.norm(trajectory[-1]-target))
```

A production DiT operates on image, video, or audio latents. Important controls include inference steps, scheduler, resolution, frame count, guidance, batching, and VAE decoding.

```python
def latent_elements(height,width,frames=1,downsample=8,channels=16):
    return (height//downsample)*(width//downsample)*frames*channels
for frames in [1,16,64]: print(frames, latent_elements(1024,1024,frames))
```
