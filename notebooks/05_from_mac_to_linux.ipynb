# From a Mac prototype to real vLLM-Omni


Your Mac is suitable for learning architecture, writing clients, validating schemas, and developing orchestration logic. Real vLLM-Omni serving should move to version-matched Linux accelerator infrastructure.

```python
from dataclasses import dataclass
@dataclass
class DeploymentPlan:
    model:str
    version:str="0.24.0"
    port:int=8091
    def commands(self):
        return [
          "uv venv --python 3.12 --seed && source .venv/bin/activate",
          f"uv pip install vllm=={self.version} --torch-backend=auto",
          "git clone --branch v%s https://github.com/vllm-project/vllm-omni.git"%self.version,
          "cd vllm-omni && uv pip install -e .",
          f"vllm serve {self.model} --omni --port {self.port}",
        ]
plan=DeploymentPlan("Tongyi-MAI/Z-Image-Turbo")
print("\n".join(plan.commands()))
```

Before deployment, verify the official release tag, model/backend support matrix, model recipe, memory requirements, API example, and health endpoint. Never mix vLLM and vLLM-Omni minor versions.
