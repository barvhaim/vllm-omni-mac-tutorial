# Serving, queues, and independent stage scaling


Disaggregation lets us scale the bottleneck instead of duplicating an entire pipeline. This deterministic simulator models a tandem queue.

```python
from omni_tutorial.serving import PipelineSimulator,StageSpec
arrivals=[i*0.5 for i in range(30)]
baseline=PipelineSimulator([StageSpec("thinker",1),StageSpec("talker",3,1),StageSpec("vocoder",.5)]).run(arrivals)
scaled=PipelineSimulator([StageSpec("thinker",1),StageSpec("talker",3,3),StageSpec("vocoder",.5)]).run(arrivals)
print("baseline makespan",baseline["makespan"])
print("scaled makespan",scaled["makespan"])
print("baseline waits",baseline["stage_stats"])
assert scaled["makespan"] < baseline["makespan"]
```

```python
import matplotlib.pyplot as plt
labels=["monolithic ratio","scaled talker"]
plt.bar(labels,[baseline["makespan"],scaled["makespan"]])
plt.ylabel("simulated time"); plt.title("Scale the bottleneck"); plt.show()
```

Real measurements need stage queue depth, service time, batch size, transfer latency, accelerator utilization, and end-to-end P50/P90/P99. These simulated numbers are not official vLLM-Omni benchmarks.
