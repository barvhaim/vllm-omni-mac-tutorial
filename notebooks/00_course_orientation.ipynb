## The four layers

```text
Application  -> voice assistant, image generator, robot agent
Server       -> HTTP, streaming, admission control, metrics
Runtime      -> scheduling, batching, memory, parallel execution
Model        -> learned weights and architecture
```

**vLLM-Omni is the runtime + serving layer.** It does not train a model or add a modality the model never learned. It makes an existing multimodal model *serve* fast and cheap.

## Why a Mac course at all?

The complete vLLM-Omni stack — heterogeneous AR/DiT stages, disaggregated serving — targets Linux + NVIDIA accelerators. You cannot run the production runtime here. But the *ideas* (paging, batching, stage graphs, queueing, guidance) are just accounting and linear algebra, and those run anywhere. This course teaches the ideas executably on your Mac, then points at the real source and shows the real commands.
