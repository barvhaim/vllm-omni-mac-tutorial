A text server usually optimizes one autoregressive loop. An omni request may traverse encoders, one or more AR models, a DiT, a VAE, or a vocoder. The serving problem is therefore a **heterogeneous graph**, not just a larger prompt.

```python
regular = ["prompt", "prefill", "decode tokens", "text"]
omni = ["multimodal input", "thinker", "talker or DiT", "vocoder or VAE", "heterogeneous output"]
print("vLLM:      ", " -> ".join(regular))
print("vLLM-Omni: ", " -> ".join(omni))
```
